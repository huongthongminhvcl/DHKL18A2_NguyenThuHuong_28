## Bài 6.1

```python
import threading
import time

def worker():
    print(f"Đang chạy luồng: {threading.current_thread().name}")
    time.sleep(1)

threads = []
for i in range(5):
    t = threading.Thread(target=worker, name=f"Thread-{i+1}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tất cả luồng đã hoàn thành.")
```

## Bài 6.2

```python
import threading
import time

def download(file):
    print(f"Đang tải {file}")
    time.sleep(2)
    print(f"Tải xong {file}")

files = ["file1.zip", "file2.zip", "file3.zip"]

threads = []
for f in files:
    t = threading.Thread(target=download, args=(f,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

## Bài 6.3

```python
import threading

def so_chan():
    for i in range(30, 51):
        if i % 2 == 0:
            print("Chẵn:", i)

def so_le():
    for i in range(30, 51):
        if i % 2 != 0:
            print("Lẻ:", i)

t1 = threading.Thread(target=so_chan)
t2 = threading.Thread(target=so_le)

t1.start()
t2.start()

t1.join()
t2.join()
```

## Bài 6.4

```python
import threading

def giai_thua(n):
    kq = 1
    for i in range(1, n + 1):
        kq *= i
    print(f"{n}! = {kq}")

numbers = [3, 5, 7, 9]

threads = []
for n in numbers:
    t = threading.Thread(target=giai_thua, args=(n,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading
import requests

def get_url(url):
    r = requests.get(url)
    print(f"{url} → {r.status_code}")

urls = [
    "https://www.google.com",
    "https://www.facebook.com",
    "https://www.yahoo.com"
]

threads = []
for url in urls:
    t = threading.Thread(target=get_url, args=(url,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

## Bài 6.7

```python
import threading
import time

def read_web(name):
    print(f"Starting {name}")
    for i in range(3):
        print(f"{name}: Web {time.strftime('%b %d %H:%M:%S %Y')}")
        time.sleep(1)
    print(f"Exiting {name}")

print("Starting Main Thread")

threads = []
for site in ["Google", "Yahoo", "Facebook"]:
    t = threading.Thread(target=read_web, args=(site,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Exiting Main Thread")
```

## Bài 6.8

```python
import threading

data = [1,3,5,7,9,2,4,6,8,10]
tong = 0
lock = threading.Lock()

def tinh_tong(arr):
    global tong
    temp = sum(arr)
    with lock:
        tong += temp

threads = []
chunk = 2
for i in range(0, len(data), chunk):
    t = threading.Thread(target=tinh_tong, args=(data[i:i+chunk],))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tổng =", tong)
```

## Bài 6.10


```python
import threading
import random

def find_max(sub_list, results, index):
    sub_max = max(sub_list)
    results[index] = sub_max
    print(f"{threading.current_thread().name} tìm max = {sub_max}")

n = 20  
data = [random.randint(0, 100) for _ in range(n)]
print("List ban đầu:", data)

num_threads = 4
chunk_size = n // num_threads

threads = []
results = [0] * num_threads

for i in range(num_threads):
    start = i * chunk_size
    end = n if i == num_threads - 1 else (i + 1) * chunk_size
    sub_list = data[start:end]

    t = threading.Thread(
        target=find_max,
        args=(sub_list, results, i),
        name=f"Thread-{i+1}"
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

max_value = max(results)
print("Giá trị lớn nhất trong list:", max_value)


```
