```python
import numpy as np

# Giả sử dữ liệu được đọc từ file vào list
heights = [180, 185, 175, 190, 170, 188] 
positions = ['GK', 'M', 'A', 'D', 'GK', 'M']

# 1. Tạo numpy array
np_positions = np.array(positions)
np_heights = np.array(heights)
print(f"Kiểu dữ liệu np_positions: {np_positions.dtype}")
print(f"Kiểu dữ liệu np_heights: {np_heights.dtype}")

# 2. Tính chiều cao trung bình của các GK
gk_heights = np_heights[np_positions == 'GK']
avg_gk = np.mean(gk_heights)
print(f"Chiều cao trung bình GK: {avg_gk}")

# 3. Tính chiều cao trung bình của các vị trí khác (không phải GK)
other_heights = np_heights[np_positions != 'GK']
avg_others = np.mean(other_heights)
print(f"Chiều cao trung bình các vị trí khác: {avg_others}")

# 4. Tạo mảng cấu trúc (Structured Array)
# U5: chuỗi tối đa 5 ký tự, f4: số thực float 32-bit
players = np.zeros(len(heights), dtype={'names':('position', 'height'), 'formats':('U5', 'f4')})
players['position'] = np_positions
players['height'] = np_heights

# 5. Sắp xếp theo height
sorted_players = np.sort(players, order='height')
print("Mảng sau khi sắp xếp:", sorted_players)
print(f"Vị trí cao nhất: {sorted_players[-1]['position']} ({sorted_players[-1]['height']})")
print(f"Vị trí thấp nhất: {sorted_players[0]['position']} ({sorted_players[0]['height']})")
```
