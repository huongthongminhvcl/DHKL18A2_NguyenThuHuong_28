```python
## 1. Tạo arr từ -9 dến 9, hiển thị phần tử & kiểu dữ liệu

import numpy as np

arr = np.arange(-9, 10)
print(arr)
print("Kiểu dữ liệu:", arr.dtype)
```

```python
##2. Tạo arr_odd chỉ chứa số lẻ

arr_odd = arr[arr % 2 != 0]
print(arr_odd)
```

```python
## 3. Tạo arr_update: số chẵn giữ nguyên, số lẻ đến 100

arr_update = np.where(arr % 2 != 0, 100, arr)
print(arr_update)
```
