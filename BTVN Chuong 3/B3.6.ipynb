```python
import numpy as np
```

```python
arr_2D = np.ones((3,3), dtype=bool)
print(arr_2D)
```

```python
arr_2D = arr_2D[:, ::-1]
arr_2D = arr_2D[::-1, :]
```

```python
print(np.diag(arr_2D))
```

```python
print(np.diag(np.fliplr(arr_2D)))
```

```python
arr_2D = arr_2D.astype(float)
arr_2D[arr_2D == arr_2D.mean()] = 0
```
