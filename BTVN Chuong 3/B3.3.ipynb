```python
import numpy as np

arr_a = [1,2,3,2,3,4,3,4,5,6]
arr_b = [7,2,10,2,7,4,9,4,9,8]
```

```python
## 1. arr_c chỉ chứa phần tử xuất hiện ở cả arr_a và arr_b

arr_c = np.intersect1d(arr_a, arr_b)
print(arr_c)
```

```python
## 2. arr_d = arr_b bỏ phần tử trùng với arr_a

arr_d = np.setdiff1d(arr_b, arr_a)
print(arr_d)
```

```python
## 3. arr_e = [2,6,1,9,10,3,27,8,6,25,16] → arr_f chia hết cho 3

arr_e = np.array([2,6,1,9,10,3,27,8,6,25,16])
arr_f = arr_e[arr_e % 3 == 0]
print(arr_f)
```
