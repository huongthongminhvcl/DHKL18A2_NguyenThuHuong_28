```python
#Chuyển đối tượng từ điển Python (sắp xếp theo khóa) thành dữ liệu JSON và in với mức thụt lề 4

import json

# Tạo một đối tượng từ điển Python
python_dict = {
    "name": "Nguyen Van A",
    "age": 21,
    "city": "Hanoi",
    "university": "UNETI"
}

# Chuyển đổi sang chuỗi JSON
# sort_keys=True → sắp xếp theo khóa (key)
# indent=4 → hiển thị với mức thụt lề 4
json_data = json.dumps(python_dict, sort_keys=True, indent=4)

# In kết quả
print("Đối tượng Python ban đầu:")
print(python_dict)

print("Sau khi chuyển sang dữ liệu JSON (sắp xếp theo khóa, thụt lề 4):")
print(json_data)
```
