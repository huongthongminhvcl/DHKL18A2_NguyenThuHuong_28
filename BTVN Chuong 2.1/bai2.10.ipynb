```python
#Viết chương trình đọc dữ liệu nhân viên từ tệp JSON và thống kê nhân viên theo đơn vị.

# Khi chạy chương trình, cần hiển thị:
# Thông tin công ty:
# Tên công ty
# Địa chỉ
# Tổng số nhân viên

# Thống kê chi tiết theo từng đơn vị:
# Tên đơn vị
# Số nhân viên trong đơn vị
# Tỷ lệ (%) của đơn vị đó so với tổng số nhân viên toàn công ty

import json

# Dữ liệu JSON (có thể lưu trong file JSON thật nếu cần)
data_json = '''
{
    "company": "Công ty TNHH Đất Việt",
    "address": "abc Giải Phóng - Hà Nội",
    "departments": [
        {"name": "Đơn vị A1", "employees": 10},
        {"name": "Đơn vị A2", "employees": 15},
        {"name": "Đơn vị A3", "employees": 20},
        {"name": "Đơn vị A4", "employees": 5},
        {"name": "Đơn vị A5", "employees": 30}

    ]
}
'''

# Chuyển từ chuỗi JSON sang đối tượng Python
data = json.loads(data_json)

# Tính tổng số nhân viên
total_employees = sum(d["employees"] for d in data["departments"])

# In thông tin công ty
print(f"Tên công ty: {data['company']}")
print(f"Địa chỉ: {data['address']}")
print(f"Tổng số nhân viên: {total_employees}")
print("----- Thống kê nhân viên theo đơn vị -----")

# Thống kê từng đơn vị
for i, dept in enumerate(data["departments"], start=1):
    name = dept["name"]
    emp = dept["employees"]
    percent = emp / total_employees * 100
    print(f"{i}. Tên đơn vị: {name}")
    print(f"   - Số nhân viên: {emp}")
    print(f"   - Tỷ lệ: {percent:.2f}%\n")
```
