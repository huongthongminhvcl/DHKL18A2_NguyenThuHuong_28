```python
#Chuyển đối tượng Python sang chuỗi JSON

import json

# Đối tượng Python
python_obj = {
    "name": "Nguyen Van A",
    "age": 21,
    "city": "Hanoi"
}

# Chuyển sang chuỗi JSON
json_data = json.dumps(python_obj)

# In ra kết quả
print("Đối tượng Python ban đầu:")
print(python_obj)

print("Sau khi chuyển sang chuỗi JSON:")
print(json_data)

# In ra các giá trị 
print("Các giá trị riêng lẻ:")
print("Tên:", python_obj["name"])
print("Tuổi:", python_obj["age"])
print("Thành phố:", python_obj["city"])


```
