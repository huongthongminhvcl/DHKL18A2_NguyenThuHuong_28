```python
# https://jsonplaceholder.typicode.com/comments?postId=1
# Hiển thị danh sách các bài post nổi bật.
# Mỗi bài hiển thị thông tin: postId, id, name, email, body.
# Giới hạn chỉ 3 bài đầu tiên.

import requests

def hien_thi_sach_noi_bat():
    # Địa chỉ API chứa các bài post nổi bật (postId = 1)
    url = "https://jsonplaceholder.typicode.com/comments?postId=1"

    # Gửi yêu cầu lấy dữ liệu
    response = requests.get(url)

    # Kiểm tra kết nối thành công
    if response.status_code == 200:
        data = response.json()
        
        print("=== DANH SÁCH CÁC BÀI POST NỔI BẬT ===")
        print(f"Tổng số bài post nổi bật: {len(data)}\n")
        
        # Chỉ hiển thị 3 bài đầu tiên
        for item in data[:3]:
            print(f"PostID: {item['postId']}")
            print(f"ID: {item['id']}")
            print(f"Tên người đăng: {item['name']}")
            print(f"Email: {item['email']}")
            print(f"Nội dung: {item['body']}\n")
    else:
        print("Không thể kết nối đến API!")

# Gọi hàm
hien_thi_sach_noi_bat()
```
