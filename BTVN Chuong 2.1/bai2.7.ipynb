```python
#Chuyển dữ liệu JSON thành đối tượng Python

import json

# Chuỗi JSON mẫu
json_data = '{"name": "Nguyen Van A", "age": 21, "city": "Hanoi"}'

# Chuyển JSON thành đối tượng Python
python_obj = json.loads(json_data)

# In ra kết quả
print("Dữ liệu JSON ban đầu:")
print(json_data)

print("Sau khi chuyển thành đối tượng Python:")
print(python_obj)

# In ra các giá trị 
print("Các giá trị riêng lẻ:")
print("Tên:", python_obj["name"])
print("Tuổi:", python_obj["age"])
print("Thành phố:", python_obj["city"])
```
