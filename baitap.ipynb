```python
# BÀI 1: HÌNH CHỮ NHẬT
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def tinh_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def hien_thi(self):
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        print(f"Chu vi: {self.tinh_chu_vi()}")
        print(f"Diện tích: {self.tinh_dien_tich()}")



```

```python
print("=== BÀI 1: HÌNH CHỮ NHẬT ===")
hcn = HinhChuNhat(5, 3)
hcn.hien_thi()
```

```python
class ThiSinh:
    def __init__(self, ho_ten, toan, ly, hoa):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def hien_thi(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}")
        print(f"Tổng điểm: {self.tong_diem()}")
```

```python
print("\n=== BÀI 2: DANH SÁCH THÍ SINH ===")
danh_sach = [
    ThiSinh("Nguyen Van A", 8, 7, 6),
    ThiSinh("Tran Thi B", 5, 5, 6),
    ThiSinh("Le Van C", 9, 8, 7)
]
for ts in danh_sach:
    ts.hien_thi()
    print("-")
```

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau
    
    def kiem_tra_hop_le(self):
        return self.mau != 0
    
    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while not self.kiem_tra_hop_le():
            print("Mẫu số phải khác 0! Nhập lại.")
            self.mau = int(input("Nhập mẫu số: "))
    
    def __str__(self):
        return f"{self.tu}/{self.mau}" if self.mau != 1 else f"{self.tu}"
    
    def hien_thi(self):
        print("Phân số:", self)
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []
    
    def isEmpty(self):
        return len(self.stack) == 0
    
    def isFull(self):
        return len(self.stack) == self.size
    
    def push(self, value):
        if not self.isFull():
            self.stack.append(value)
        else:
            print("Ngăn xếp đầy, không thể thêm phần tử!")
    
    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
        else:
            print("Ngăn xếp rỗng, không thể lấy phần tử!")
            return None
```

```python
# --- Bài 4 ---
st = Stack(3)
st.push(1)
st.push(2)
st.push(3)
print("Is full?", st.isFull())
print("Pop:", st.pop())
print("Is empty?", st.isEmpty())
```

```python
class StackCount(Stack):
    def __init__(self, size):
        super().__init__(size)
    
    def count(self):
        return len(self.stack)
```

```python
stc = StackCount(5)
stc.push(10)
stc.push(20)
print("Số phần tử:", stc.count())
```
