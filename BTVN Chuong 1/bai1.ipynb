```python
# Xây dựng lớp Hình chữ nhật với:
# Thuộc tính: độ dài hai cạnh.

# Phương thức:
# Nhập dữ liệu cho hai cạnh.
# Tính chu vi.
# Tính diện tích.
# In thông tin (hai cạnh, chu vi, diện tích) ra màn hình.
# Viết chương trình chính cho phép nhập dữ liệu của một hình chữ nhật và in ra các thông tin trên

class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.dai = dai
        self.rong = rong

    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        chu_vi = 2 * (self.dai + self.rong)
        print(f"Chu vi: {chu_vi}")

    def tinh_dien_tich(self):
        dien_tich = self.dai * self.rong
        print(f"Diện tích: {dien_tich}")

    def in_thong_tin(self):
        print("----- Thông tin hình chữ nhật -----")
        print(f"Chiều dài: {self.dai}")
        print(f"Chiều rộng: {self.rong}")
        self.tinh_chu_vi()
        self.tinh_dien_tich()

if __name__ == "__main__":
    hcn = HinhChuNhat()
    hcn.nhap()
    hcn.in_thong_tin()
```
