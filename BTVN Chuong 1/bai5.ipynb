```python
# Hiệu chỉnh lớp Stack bằng cách thêm vào thành phần count để trả về số phần tử trên ngăn xếp.

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            print("Stack rỗng!")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            print("Stack rỗng!")

    def is_empty(self):
        return len(self.items) == 0

    def count(self):
        return len(self.items)


s = Stack()
s.push(10)
s.push(20)
s.push(30)

print("Số phần tử trong stack:", s.count())  

s.pop()
print("Số phần tử trong stack:", s.count())
```
