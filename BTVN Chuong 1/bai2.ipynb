```python
# Bài 2: Để quản lý điểm thi vào một trường đại học, hãy xây dựng lớp ThiSinh để mô tả các thí sinh với:
# Thuộc tính: Họ tên thí sinh; Điểm Toán, Lý, Hóa.

# Phương thức:
# Nhập thông tin thí sinh
# In thông tin thí sinh
# Tính tổng điểm của thí sinh
# Trên cơ sở lớp trên, viết chương trình thực hiện các yêu cầu:
# Nhập họ tên và điểm của một danh sách thí sinh.
# Đưa ra màn hình danh sách thí sinh trúng tuyển (giả sử điểm chuẩn vào trường là 20), theo thứ tự điểm từ cao xuống thấp.
```

```python
class ThiSinh:
    def __init__(self, ho_ten, toan, ly, hoa):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def __str__(self):
        return f"{self.ho_ten} - Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}, Tổng: {self.tong_diem()}"

def lay_tong_diem(thisinh):
    return thisinh.tong_diem()

ds_thisinh = []

n = int(input("Nhập số lượng thí sinh: "))
for i in range(n):
    ho_ten = input(f"\nNhập họ tên thí sinh {i+1}: ")
    toan = int(input("Điểm Toán: "))
    ly = int(input("Điểm Lý: "))
    hoa = int(input("Điểm Hóa: "))
    ts = ThiSinh(ho_ten, toan, ly, hoa)
    ds_thisinh.append(ts)

ds_thisinh.sort(key=lay_tong_diem, reverse=True)

print("\nDanh sách thí sinh trúng tuyển (điểm chuẩn = 20):")
for ts in ds_thisinh:
    if ts.tong_diem() >= 20:
        print(ts)
```
