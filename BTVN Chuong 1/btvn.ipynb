Câu 1 — So sánh OOP và POP

Đơn vị cấu trúc: OOP: object (dữ liệu + phương thức); POP: procedure/function (hàm xử lý dữ liệu).

Tổ chức dữ liệu: OOP đóng gói dữ liệu với thao tác; POP tách rời dữ liệu và hàm (thường dùng biến toàn cục).

Khả năng tái sử dụng & mở rộng: OOP hỗ trợ kế thừa và đa hình → dễ tái sử dụng, mở rộng; POP khó quản lý khi hệ thống lớn.

Mô hình hóa thực tế: OOP dễ mô phỏng đối tượng thực thế; POP phù hợp cho chương trình đơn giản, thuật toán rõ ràng.

Ví dụ ngôn ngữ: OOP: Java, C++ (hỗ trợ OOP); POP: C, Pascal (theo phong cách cấu trúc).

Câu 2 — Tính chất đặc thù của lập trình hướng đối tượng

Đóng gói (Encapsulation) — gom dữ liệu và phương thức, dùng access modifiers (private/protected/public).

Kế thừa (Inheritance) — lớp con kế thừa thuộc tính/phương thức của lớp cha.

Đa hình (Polymorphism) — cùng một gọi phương thức nhưng thực thi khác nhau (overload/override).

Trừu tượng (Abstraction) — ẩn chi tiết hiện thực, chỉ expose giao diện cần thiết.

(bổ sung) Ràng buộc thời gian chạy / binding động và giao tiếp giữa đối tượng.

Câu 3 — Tính đóng gói (encapsulation) có đặc trưng gì?

Định nghĩa ngắn: Đóng gói là gom dữ liệu (thuộc tính) và thao tác trên dữ liệu (phương thức) vào một đơn vị (class) và kiểm soát truy cập.

Đặc trưng: có chế độ truy cập (private/protected/public), ẩn thông tin (information hiding), chỉ cho phép thao tác qua phương thức công khai.

Lợi ích: bảo vệ trạng thái đối tượng, giảm phụ thuộc (coupling), dễ bảo trì, kiểm soát hợp lệ dữ liệu.

Câu 4 — Các loại kế thừa trong OOP

Single inheritance: lớp con kế thừa từ một lớp cha duy nhất.

Multiple inheritance: lớp con kế thừa từ nhiều lớp cha (C++ hỗ trợ; Java thì dùng interface thay thế).

Multilevel inheritance: chuỗi kế thừa: lớp A → B → C (ông → cha → con).

Hierarchical inheritance: một lớp cha có nhiều lớp con.

Hybrid (hoặc multipath): kết hợp các kiểu trên (ví dụ multilevel + multiple).
(Lưu ý: một số ngôn ngữ giới hạn kiểu kế thừa; dùng interface/trait để thay thế.)

Câu 5 — Đa hình (Polymorphism) là gì?

Định nghĩa: Khả năng sử dụng cùng một tên phương thức/giao diện để gọi các cách thực thi khác nhau tùy loại đối tượng.

Phân loại: Compile-time (static) — nạp chồng hàm/overloading; Runtime (dynamic) — ghi đè/overriding với binding động (virtual methods).

Ví dụ ngắn: shape.draw() — đối với Circle gọi draw của Circle, với Square gọi draw của Square.

Lợi ích: viết mã linh hoạt, mở rộng dễ dàng, giảm if/else kiểm loại.

Câu 6 — Khái niệm trừu tượng (abstraction)

Định nghĩa: Trừu tượng là rút gọn, chỉ trình bày những thuộc tính và hành vi cần thiết của đối tượng, ẩn đi chi tiết triển khai.

Cách hiện thực: dùng abstract class hoặc interface để khai báo giao diện/method mà không nêu chi tiết.

Lợi ích: giảm phức tạp, tập trung vào chức năng chính, dễ thay đổi hiện thực mà không ảnh hưởng người dùng giao diện.
