```python
# Bài 3. Xây dựng lớp Phân số (PS) gồm:
# Thuộc tính:
# Tử số
# Mẫu số

# Phương thức:
# Kiểm tra tính hợp lệ của phân số (mẫu số ≠ 0)
# Nhập phân số từ bàn phím
# In phân số ra màn hình

class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau

    def hop_le(self):
        return self.mau != 0

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))

    def xuat(self):
        if self.hop_le():
            print(f"Phân số: {self.tu}/{self.mau}")
        else:
            print("Phân số không hợp lệ (mẫu số = 0).")

ps = PhanSo()
ps.nhap()
ps.xuat()
```
